# Visual comparison of MSE and log loss functions

```python
import numpy as np

b = 0
w = np.arange(-20,20.1,.1)

sigmoid = lambda w : 1 / (1 + np.e**(-w+b))

yhat = sigmoid(w)
```

```python

def f_mse(y, yhat) :
    return (yhat - y)**2

def f_sigmoid(y, yhat) :    
    return - (y * np.log(yhat) + (1-y) * np.log(1 - yhat))


def plot_loss(w,l0, l1, title = 'Loss') :
    plt.plot(w, l0, '-o', w, l1, '-o')
    plt.ylabel('Loss')
    plt.xlabel('w')
    plt.title(title)
```

```python
# mse losses
L0_mse = f_mse(0,yhat)
L1_mse = f_mse(1,yhat)

# sigmoid losses
L0_sigmoid = f_sigmoid(0,yhat)
L1_sigmoid = f_sigmoid(1,yhat)

plt.figure().set_figwidth(15)
plt.subplot(121)
plot_loss(w,L0_mse,L1_mse, 'MSE loss')

plt.subplot(122)
plot_loss(w,L0_sigmoid,L1_sigmoid, 'Log loss')
```
